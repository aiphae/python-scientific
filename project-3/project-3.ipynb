```python
import string
import requests
import numpy
import matplotlib.pyplot as plt
```

```python
book = requests.get('https://www.gutenberg.org/files/35/35-0.txt')
text = book.text

to_replace = [
    '\r\n' # New line
]

for str_ in to_replace:
    text = text.replace(str_, " ")

words = text.split(" ")
max_len = max(len(word) for word in words)

plt.hist([len(w) for w in words], bins=range(1, max_len + 2))
plt.show()
```

```python
letters = [ch.lower() for ch in text if ch in string.ascii_lowercase]
plt.hist(letters, bins=26)
plt.show()
```

```python
values, counts = numpy.unique(letters, return_counts=True)
probs = counts / counts.sum()

entropy_letters = -numpy.sum(probs * numpy.log2(probs))
print(entropy_letters)

plt.bar(values, -probs * numpy.log2(probs))
plt.show()
```
